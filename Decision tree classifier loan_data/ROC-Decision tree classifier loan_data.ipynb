# Train Test Split


```python
from sklearn.model_selection import train_test_split
```

```python
X = final_data.drop('not.fully.paid',axis=1)
y = final_data['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

# Training a Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
```

# Predictions and Evaluation of Decision Tree

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
cm = confusion_matrix(y_test,predictions)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test,predictions))
```

# Random forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test, rfc_pred))
```

```python
print(classification_report(y_test, rfc_pred))
```

# ROC

```python
r_probs = [0 for _ in range(len(y_test))]
dtree_probs = dtree.predict_proba(X_test)
rfc_probs = rfc.predict_proba(X_test)
```

```python
dtree_probs = dtree_probs[:, 1]
rfc_probs = rfc_probs[:, 1]
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
r_auc = roc_auc_score(y_test, r_probs)
dtree_auc = roc_auc_score(y_test, dtree_probs)
rfc_auc = roc_auc_score(y_test, rfc_probs)
```

```python
print('Random (chance) Prediction: AUROC = %.3f' % (r_auc))
print('Decision Tree Classifier: AUROC = %.3f' % (dtree_auc))
print('Random Forest: AUROC = %.3f' % (rfc_auc))
```

```python
r_fpr, r_tpr, _ = roc_curve(y_test, r_probs)
dtree_fpr, dtree_tpr, _ = roc_curve(y_test, dtree_probs)
rfc_fpr, rfc_tpr, _ = roc_curve(y_test, rfc_probs)
```

```python
plt.plot(r_fpr, r_tpr, linestyle='--', label='Random prediction (AUROC = %0.3f)' % r_auc)
plt.plot(dtree_fpr, dtree_tpr, marker='.', label='Decision Tree Classifier (AUROC = %0.3f)' % dtree_auc)
plt.plot(rfc_fpr, rfc_tpr, marker='.', label='Random Forest (AUROC = %0.3f)' % rfc_auc)

# Title
plt.title('ROC Plot')
# Axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# Show legend
plt.legend() # 
# Show plot
plt.show()
```
