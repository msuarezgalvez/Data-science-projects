# Game of thrones battle analysis

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
battle = pd.read_csv('battles.csv')
```

```python
battle.head()
```

```python
battle.shape
```

```python
battle.info()
```

```python
battle.rename(columns={'attacker_1':'primary_attacker'},inplace=True)
battle.head()
```

```python
battle.rename(columns={'defender_1':'primary_defender'},inplace=True)
battle.head()
```

```python
battle['attacker_king'].value_counts()
```

```python
battle['location'].value_counts()
```

```python
sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(x='attacker_king',y='attacker_size',data=battle)
plt.xlabel("Attacker King")
plt.ylabel("Attacker Size")
plt.title("Attacker Size v/s Attacker King")
plt.show()
```

```python
sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(x='defender_king',y='defender_size',data=battle)
plt.xlabel("Defender King")
plt.ylabel("Defender Size")
plt.title("Defender Size v/s Defender King")
plt.show()
```

```python
sns.countplot(x=battle['attacker_king'],hue=battle['battle_type'])
plt.xlabel("Attacker King")
plt.ylabel("Count")
plt.title("Battle Type count v/s Attacker king")
plt.show()
```
